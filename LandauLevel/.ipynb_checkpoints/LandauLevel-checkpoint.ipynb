```python
import matplotlib.pyplot as plt
import numpy as np
##
N = 15; ## the number of subbands of each band
##
def MagHamil_Weyl(N,B,k3):
    MagHamil = np.matrix(np.zeros((N*2,N*2)))
    for ii in np.arange(0,N,1):
            Hn = np.matrix([[k3,np.sqrt(B*(ii+1))],[np.sqrt(B*(ii+1)),-k3]])
            MagHamil[ii*2:ii*2+2,ii*2:ii*2+2] = Hn
    return[MagHamil]

def MagHamil_NL_perp(N,B,k3,m):
    MagHamil = np.matrix(np.zeros((N*2,N*2)))
    for ii in np.arange(0,N,1):
            Hn = np.matrix([[k3,m-B*ii],[m-B*ii,-k3]])
            MagHamil[ii*2:ii*2+2,ii*2:ii*2+2] = Hn
    return[MagHamil]
##
def MagHamil_NL_para(N,B,k3,m):
    MagHamil = np.matrix(np.zeros((N*2,N*2)))
    for ii in np.arange(0,N,1):
            H00 = np.matrix([[0,1],[1,0]])*(m-k3**2-B*(ii+1)/2);
            MagHamil[ii*2:ii*2+2,ii*2:ii*2+2] = H00;
    for ii in np.arange(0,N-1,1):
            H01 = np.matrix([[1,0],[0,1]])*(np.sqrt(B*(ii+1)/2));
            MagHamil[ii*2:ii*2+2,ii*2+2:ii*2+4] = H01;
            H01t = np.matrix([[1,0],[0,1]])*(np.sqrt(B*(ii+1)/2));
            MagHamil[ii*2+2:ii*2+4,ii*2:ii*2+2] = H01t;
    for ii in np.arange(0,N-2,1):
            H02 = np.matrix([[0,1],[1,0]])*(-(B/2)*np.sqrt((ii+2)*(ii+1)));
            MagHamil[ii*2:ii*2+2,ii*2+4:ii*2+6] = H02;
            MagHamil[ii*2+4:ii*2+6,ii*2:ii*2+2] = H02;
    return[MagHamil]
##
def LandauLevel(N,B_list,k3_list,m,ind_Hamil):
    nB = B_list.size;  nk3 = k3_list.size;
    if nB > 1:
        nPara = nB
    else:
        nPara = nk3
    ##
    LL = np.matrix(np.zeros((nPara,N*2)))
    for jj in np.arange(1,nPara,1):
        if nB > 1:
            B = B_list[jj-1]
            k3 = k3_list[0]
        else:
            k3 = k3_list[jj-1]
            B = B_list[0]
        if ind_Hamil == 1:
            MagHamil = MagHamil_Weyl(N,B,k3)
        elif ind_Hamil == 2:
            MagHamil = MagHamil_NL_perp(N,B,k3,m)
        elif ind_Hamil == 3:
            MagHamil = MagHamil_NL_para(N,B,k3,m)
        ##
        D,v = np.linalg.eig(MagHamil)
        D.shape = (N*2,1)
        LL[jj,:] = np.transpose(np.sort(D,axis=0))[0,:]
    return[LL]
```

```python
lw = 1;  cl = [0,0,0];  fs = 16;
##------
def set_plt(xlab,ylab,axis):
    ft1 = {'family' : 'Times New Roman',
       'weight' : 'normal',
       'size'   : fs,}
    plt.xlabel(xlab,ft1); plt.ylabel(ylab,ft1);
    plt.axis(axis)
    plt.xticks(np.arange(axis[0],axis[1]+0.01,(axis[1]-axis[0])/4),fontsize=fs); 
    plt.yticks(np.arange(axis[2],axis[3]+0.01,(axis[3]-axis[2])/4),fontsize=fs)
    return[]
```

## 0 Methods and Procedure

The key principle of the computation method is that we can express the momentum operators in terms of the creation and annihilation operators, say:
$$P_{x}=\sqrt{\frac{\hbar eB}{2}}(a^{\dagger}+a),~~P_{y}=-i\sqrt{\frac{\hbar eB}{2}}(a^{\dagger}-a),$$

where $\vec{P}=-i\hbar\nabla + e\vec{A}$. Or equivalently,
$$k_{x}=\sqrt{\frac{eB}{2\hbar}}(a^{\dagger}+a),~~k_{y}=-i\sqrt{\frac{eB}{2\hbar}}(a^{\dagger}-a),$$

and
$$k_{+}=\sqrt{\frac{2eB}{\hbar}}a^{\dagger},~~k_{-}=\sqrt{\frac{2eB}{\hbar}}a.$$

In the natural unit, we have
$$k_{+}=\sqrt{2B}a^{\dagger},~~k_{-}=\sqrt{2B}a.$$

## 1 Weyl SMï¼š $\vec{B}=B\hat{e}_{3}$

Start with a continum Hamiltonian of a Weyl point carrying a postive chirality:

$$\tag{1}\label{ham_weyl}
\hat{H}_{W}=\begin{pmatrix}k_{3}&k_{-}\\k_{+}&-k_{3}\end{pmatrix}
\rightarrow
\hat{H}_{W}^{LL}(k_{3})=\begin{pmatrix}k_{3}&\sqrt{B}a\\\dagger&-k_{3}\end{pmatrix},$$

which is under the basis $(|\varphi_{1}\rangle,|\varphi_{2}\rangle)^{T}$ and $k_{\pm}=k_{1}\pm ik_{y}$.

- After Landau level quantization, there is a set of LL basis:

$(|\varphi_{2},0\rangle,|\varphi_{1},0\rangle,|\varphi_{2},1\rangle,|\varphi_{1},1\rangle,|\varphi_{2},2\rangle,\cdots,|\varphi_{1},N\rangle)^{T}
\equiv (|\varphi_{2},0\rangle,|\psi_{0}\rangle,|\psi_{1}\rangle,\cdots,|\varphi_{1},N\rangle)^{T}$

- and the magnetic Hamiltonian is

$$\tag{2}\label{mag_ham}
H_{W}^{LL}(k_{3})=\begin{pmatrix}
-k_{3}& 0 &0&\cdots&0\\
&\langle \psi_{0}|\hat{H}|\psi_{0}\rangle&0& \cdots&0\\
&&\langle \psi_{1}|\hat{H}|\psi_{1}\rangle& \cdots&0\\
&\dagger&&\ddots&0\\
&&&&k_{3}\\
\end{pmatrix}$
$\leadsto E_{n}=\pm\sqrt{Bn+k_{3}^{2}}(n\geqslant 0)$$
   
where
$\langle \psi_{n}|\hat{H}|\psi_{n}\rangle 
=\begin{pmatrix}k_{3}&\sqrt{B}\sqrt{n+1}\\\dagger&-k_{3}
\end{pmatrix}~\leadsto E_{n}=\pm\sqrt{B(n+1)+k_{3}^{2}}~~(n\geqslant 0)$

$\langle\varphi_{2},0|-k_{3}|\varphi_{2},0\rangle=-k_{3}$

$\langle\varphi_{1},N|k_{3}|\varphi_{1},N\rangle=k_{3}$

```python
ind_Hamil = 1
##
B_list = np.arange(0,10,0.01);
k3_list = np.array([0]);
LL = LandauLevel(N,B_list,k3_list,0,ind_Hamil)[0];
##
for ii in np.arange(0,N*2-1,1):
    plt.plot(B_list,LL[:,ii],color=cl,linewidth=lw)
##
set_plt('$B (T)$','$E$',[0,10,-10,10])
plt.text(0.5,8,'$k_{3}=0$',fontsize=fs,color=[1,0,0])
plt.show()
plt.savefig('output/LL_Weyl_B.png')
```

```python
k3_list = np.arange(-1,1,0.001)
B_list = np.array([2])
LL = LandauLevel(N,B_list,k3_list,0,ind_Hamil)[0];
##
for ii in np.arange(1,N*2,1):
    plt.plot(k3_list,LL[:,ii-1],color=cl,linewidth=lw)
##
set_plt('$k_{3}$','$E$',[-1,1,-8,8])
plt.text(-0.9,6,'$B=2$',fontsize=fs,color=[1,0,0])
plt.show()
```

## 2 Nodal-line SM with a Ring lying in the $k_{1}-k_{2}$ plane

Start with a continum Hamiltonian of a generic two-band nodal-ring semimetal:

$$\tag{1}\label{ham_weyl}
\hat{H}_{NL}=\begin{pmatrix}k_{3}&m-(k_{1}^{2}+k_{2}^{2})\\ \dagger&-k_{3}\end{pmatrix}
=\begin{pmatrix}k_{3}&m-k_{+}k_{-}\\\dagger&-k_{3}\end{pmatrix}$$

which is under the basis $(|\varphi_{1}\rangle,|\varphi_{2}\rangle)^{T}$ and $k_{\pm}=k_{1}\pm ik_{y}$.

### 2.1  $\vec{B}=B\hat{e}_{3}$: Perpenticular to the Ring

$\hat{H}_{NL}
\rightarrow
\hat{H}^{LL}_{NL}(k_{3})=\begin{pmatrix}k_{3}&m-Ba^{\dagger}a\\\dagger&-k_{3}\end{pmatrix}$

 with basis $(|\varphi_{1}\rangle,|\varphi_{2}\rangle)^{T}$

- New basis after Landau level quantization:

$(|\varphi_{1},0\rangle,|\varphi_{2},0\rangle,|\varphi_{1},1\rangle,|\varphi_{2},1\rangle,\cdots)^{T}
\equiv (|\psi_{0}\rangle,|\psi_{1}\rangle,\cdots)^{T}$

- Magnetic Hamiltonian

$\langle \psi_{n}|\hat{H}|\psi_{n}\rangle 
=\begin{pmatrix}k_{3}&m-Bn\\\dagger&-k_{3}
\end{pmatrix}$

$H_{LL}(k_{3})=\begin{pmatrix}
\langle \psi_{0}|\hat{H}|\psi_{0}\rangle&0& \cdots\\
&\langle \psi_{1}|\hat{H}|\psi_{1}\rangle& \cdots\\
\dagger&&\ddots\\
\end{pmatrix}$
    $\leadsto E_{n,\pm}=\pm\sqrt{\left(m-Bn\right)^{2}+k_{3}^{2}}~(n\geqslant 0)$

```python
ind_Hamil = 2;  m=2
##
B_list = np.arange(0,2,0.001)
k3_list = np.array([0.5])
LL = LandauLevel(N,B_list,k3_list,m,ind_Hamil)[0];
##
for ii in np.arange(0,N*2-1,1):
    plt.plot(B_list,LL[:,ii],color=cl,linewidth=lw)
##
set_plt('$B (T)$','$E$',[0,2,-10,10])
plt.text(0.1,8,'$k_{3}=0.5$',fontsize=fs,color=[1,0,0])
plt.show()
plt.savefig('output/LL_NL_perp_B.png')
```

```python
k3_list = np.arange(-1,1,0.001)
B_list = np.array([3])
LL = LandauLevel(N,B_list,k3_list,m,ind_Hamil)[0]
##
for ii in np.arange(1,N*2,1):
    plt.plot(k3_list,LL[:,ii-1],color=cl,linewidth=lw)
##
set_plt('$k_{3}$','$E$',[-1,1,-10,10])
plt.text(-0.8,6,'$B=2.5T$',fontsize=fs,color=[1,0,0])
plt.show()
```

### 2.2 $\vec{B}=B\hat{e}_{1}$: Parallel to the Ring

$\hat{H}_{NL}
\rightarrow
\hat{H}_{NL}^{LL}(k_{1})=\begin{pmatrix}\frac{\sqrt{B}}{\sqrt{2}i}(a^{\dagger}-a)&m-k_{1}^{2}-\frac{B}{2}(a^{\dagger}+a)^{2}\\\dagger&-\frac{\sqrt{B}}{\sqrt{2}i}(a^{\dagger}-a)\end{pmatrix}$

with basis $(|\varphi_{1}\rangle,|\varphi_{2}\rangle)^{T}$

- After Landau level quantization, there is a set of LL basis:

$(|\varphi_{1},0\rangle,|\varphi_{2},0\rangle,|\varphi_{1},1\rangle,|\varphi_{2},1\rangle,\cdots)^{T}
\equiv (|\psi_{0}\rangle,|\psi_{1}\rangle,\cdots)^{T}$

- Magnetic Hamiltonian

$\langle\psi_{n}|\hat{H}|\psi_{n}\rangle =\begin{pmatrix}0&m-k_{1}^{2}-\frac{B}{2}(2n+1)\\ \dagger&0\end{pmatrix}$

$\langle\psi_{n}|\hat{H}|\psi_{n+1}\rangle =\begin{pmatrix}-\frac{\sqrt{B}}{\sqrt{2}i}\sqrt{n+1}&0\\ 0&-\frac{\sqrt{B}}{\sqrt{2}i}\sqrt{n+1}\end{pmatrix}$

$\langle\psi_{n}|\hat{H}|\psi_{n+2}\rangle =\begin{pmatrix}0&-\frac{B}{2}\sqrt{(n+2)(n+1)}\\ -\frac{B}{2}\sqrt{(n+2)(n+1)}&0\end{pmatrix}$

- When the strength of the magtnetic field is strong enough, we pick up the two-dimensional Hilbert space of the zeroth LL, up to the zeroth approximation:

$$H_{LL}^{(0)}(k_{1})=\begin{pmatrix}0&m-k_{1}^{2}-\frac{B}{2}\\ \dagger&0\end{pmatrix}$$

Expanding it around two different wave vector $k_{1}=\pm Q$:

$$H_{LL}^{(0)}(Q+\delta k_{1})=\left[\left(m-\frac{B}{2}+Q^{2}\right)-2Q\delta k_{1}\right]\sigma_{x}+O(\delta k_{1})\rightarrow H_{LL}^{Q}(\delta k_{1})\equiv -2Q\delta k_{1}\sigma_{1}$$

$$H_{LL}^{(0)}(-Q+\delta k_{1})=\left[\left(m-\frac{B}{2}+Q^{2}\right)+2Q\delta k_{1}\right]\sigma_{x}+O(\delta k_{1})\rightarrow H_{LL}^{-Q}(\delta k_{1})\equiv 2Q\delta k_{1}\sigma_{1}$$

- Chirality: A given massless particle appears to spin in the same direction along its axis of motion regardless of point of view of the observer. Following this, the chirality of such two effective Hamiltonian at $\pm Q$ carry opposite chiralities.

```python
ind_Hamil = 3;  m=2
##
B_list = np.arange(0,1,0.001);
k1_list = np.array([0.5]);
LL = LandauLevel(N,B_list,k3_list,m,ind_Hamil)[0];
##
for ii in np.arange(0,N*2-1,1):
    plt.plot(B_list,LL[:,ii],color=cl,linewidth=lw)
##
set_plt('$B (T)$','$E$',[0,1,-5,5])
plt.text(0.05,4,'$k_{3}=0.5$',fontsize=fs,color=[1,0,0])
plt.show()
plt.savefig('output/LL_NL_para_B.png')
```

```python
k1_list = np.arange(-1,1,0.001);
B_list = np.array([8]);
LL = LandauLevel(N,B_list,k3_list,m,ind_Hamil)[0];
##
for ii in np.arange(1,N*2-1,1):
    plt.plot(k1_list,LL[:,ii-1],color=cl,linewidth=lw)
##
set_plt('$k_{3}$','$E$',[-1,1,-5,5])
plt.text(-0.8,4,'$B=8T$',fontsize=fs,color=[1,0,0])
plt.show()
```
